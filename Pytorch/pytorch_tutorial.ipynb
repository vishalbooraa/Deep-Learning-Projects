## Image tensor format [B,C,H,W] and reshaping

```python
#B= Batch Size
#C= Channels
#H= Height of image
#W= Width of image

a=torch.randn(8,3,256,256)

b=a.reshape(8,3,256*256)

c=a.permute(0,3,2,1)

print("size of a",a.size())
print("size of b",b.size())
print("size of c",c.size())
```

## Autograd computational graph, .backward(), .grad, .no_grad, detach()

```python
a=torch.tensor(2.0, requires_grad=True)
b=torch.tensor(3.0, requires_grad=True)

z=a**2 + b*3
out=2.0*z

out.backward()
print("out:",out)
```

```python
print(a.grad)
print(b.grad)
```

```python
z=a**2 + b*3
out=2.0*z

out.backward()
print("out:",out)
print(a.grad)
print(b.grad)
```

```python
with torch.no_grad():
    pred=a**2+3*b
    print(pred)
```

## nn.Module,layers(Linear,Covo2d), and feature maps

```python
import torch.nn as nn
```

```python
conv= nn.Conv2d(in_channels=3,out_channels=64,kernel_size=3,padding=1)
x=torch.randn(8,3,256,256)
out=conv(x)
print("input shape:",x.size())
print("output shape:",out.size())
```

```python
class CNNNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net=nn.Sequential(
            nn.Conv2d(3,16,3,padding=1),
            nn.ReLU(),
            nn.Conv2d(16,8,3,padding=1),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(8*256*256,10)
        )

    def forward(self,x):
        return self.net(x)
```

```python
model=CNNNet()
print(model)
```

```python
x=torch.randn(8,3,256,256)
out=model(x)
```

```python
print(out)
```

```python
print(out.size())
```
